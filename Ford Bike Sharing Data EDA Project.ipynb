# Data Cleaning

```python
# Dropping rows with missing values for simplicity
df_clean = df.dropna()

# Confirm no missing data
print("\n--- After Cleaning: Missing Values ---")
print(df_clean.isnull().sum())
```

```python
# Standardize station names (lowercase, remove spaces)
df['start_station_name'] = df['start_station_name'].str.lower().str.strip()
df['end_station_name'] = df['end_station_name'].str.lower().str.strip()

# Drop rows where start or end station ID is missing
df = df.dropna(subset=['start_station_id', 'end_station_id'])

# Fix unrealistic birth years
df = df[(df['member_birth_year'] >= 1900) & (df['member_birth_year'] <= 2010)]

# (Optional) Reset index after dropping
df = df.reset_index(drop=True)
```

## 3. ***Data Wrangling***

### Data Wrangling Code

```python
# Converting start_time and end_time to datetime format
df_clean['start_time'] = pd.to_datetime(df_clean['start_time'])
df_clean['end_time'] = pd.to_datetime(df_clean['end_time'])
```

```python
# Extracting additional time-based features
df_clean['start_hour'] = df_clean['start_time'].dt.hour
df_clean['start_day'] = df_clean['start_time'].dt.day_name()
df_clean['trip_duration_minutes'] = df_clean['duration_sec'] / 60
```

### What all manipulations have you done and insights you found?

Data Cleaning and Manipulation Done:

Checked Unique Values:

Inspected unique values for each column.

Displayed actual values where the number of unique values was  10.

Handled Missing Data:

Dropped rows containing any missing (NaN) values for simplicity and consistency.

Standardized Text Columns:

Converted start_station_name and end_station_name to lowercase and stripped extra spaces for uniformity.

Filtered Out Unrealistic Data:

Dropped rows where start_station_id or end_station_id was missing.

Removed rows with unrealistic member_birth_year values (kept years between 1900 and 2010 only).

Reset Index:

Reset the DataFrame index after removing unwanted rows for a clean structure.

Datetime Conversion:

Converted start_time and end_time columns into proper datetime objects.

Feature Engineering:

Extracted start hour (start_hour) from start_time.

Extracted day name (start_day) from start_time.

Created a new column for trip duration in minutes (trip_duration_minutes) by dividing duration_sec by 60.

## ***4. Data Vizualization, Storytelling & Experimenting with charts : Understand the relationships between variables***

#### Chart - 1

```python
#  Distribution of trip durations
plt.figure(figsize=(10,6))
sns.histplot(df_clean['trip_duration_minutes'], bins=100, kde=True)
plt.xlim(0, 60)
plt.title('Distribution of Trip Durations (0-60 min)', fontsize=16)
plt.xlabel('Trip Duration (minutes)')
plt.ylabel('Count')
plt.show()
```

##### 1. Why did you pick the specific chart?

I selected a histogram with a KDE (Kernel Density Estimate) plot to explore the distribution of trip durations. Histograms are ideal for showing the frequency of continuous numerical data, while the KDE overlay provides a smooth curve that reveals the underlying pattern in the data. Since trip duration is a continuous variable measured in minutes, this approach effectively captures both the spread and the common values in one plot. Additionally, setting the x-axis to a maximum of 60 minutes helps focus the analysis on typical trips, avoiding distortion from extreme outliers.

##### 2. What is/are the insight(s) found from the chart?

Majority of trips are clustered around the 5-20 minute range.

There is a sharp peak in the lower duration values, which indicates that most trips are relatively short.

After about 30 minutes, the number of trips drops significantly, confirming that longer trips are rare.

There are some outliers where trips last close to 60 minutes, but these are minimal compared to the shorter trips.

The KDE curve shows a right-skewed distribution, meaning most trips are short with a tail extending toward longer durations.



##### 3. Will the gained insights help creating a positive business impact?


Understanding that the majority of trips are short allows the business to:

Optimize bike rebalancing schedules and ensure that bikes are available at high-demand locations quickly.

Design pricing models  for example, introduce short trip promotions or passes for trips under 30 minutes.
